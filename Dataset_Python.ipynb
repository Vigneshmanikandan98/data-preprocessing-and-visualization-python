```python
# 5 points
# Task 2-a: Find out how many doctorates there are that are married.

# 10 points
# Task 2-b: Retrieve and print the records of Americans over 40 years who have been putting in more than 40 hours a week of work.

# 5 points
# Task 2-c: Discover the top 10 nations with the highest incomes and print a list of them.

# 10 points
# Task 2-d: Fetch and print the records of people having doctoral degrees and are divorced.  
```

```python
# Task 2-a: To find out how many doctorates there are that are married we iterated through the data and coounted them using a counter.
df_data.columns = [i.replace('-', '_') for i in df_data.columns]

i=0
for index, row in df_data.iterrows():
    if row["education"] == 'Doctorate': 
      if row["marital_status"] == 'Married-civ-spouse' or row["marital_status"] == 'Married-spouse-absent' or row["marital_status"] == 'Married-AF-spouse':
        i = i+1

print('The Total number of married doctorates are: ',i)
      
```

```python
# Task 2-b: To retrieve and print the records of Americans over 40 years who have been putting in more than 40 hours a week of work we created a new data frame and 
# added the rows which satisfies the condition and printed it along with the total count.
df_data.columns = [i.replace('-', '_') for i in df_data.columns]
df2 = pd.DataFrame(columns=df_data.columns)
i=0
for index, row in df_data.iterrows():
    if row["native_country"] == 'United-States': 
      if row["age"] > 40 and row["hours_per_week"] > 40 :
        i = i+1
        df2.loc[len(df2.index)] = row

print('The Total number of Americans over 40 years who have been putting in more than 40 hours a week of work are: ',i)
print(df2)
      
```

```python
# Task 2-c: To discover the top 10 nations with the highest incomes and we used groupby() to group the countries and added the total capital gains income of the country and printed the top 10 countries.
df_data.columns = [i.replace('-', '_') for i in df_data.columns]

df2=df_data.groupby('native_country')['capital_gain'].sum().nlargest(10)
print(df2)
```

```python
# Task 2-d: To retrieve and print the records of divorced doctorates we created a new data frame and 
# added the rows which satisfies the condition and printed it along with the total count.
df_data.columns = [i.replace('-', '_') for i in df_data.columns]
df2 = pd.DataFrame(columns=df_data.columns)
i=0
for index, row in df_data.iterrows():
    if row["education"] == 'Doctorate': 
      if row["marital_status"] == 'Divorced':
        i = i+1
        df2.loc[len(df2.index)] = row

print('The Total number of divorced doctorates are: ',i) 
print(df2)
```

```python
# Task 3-b: This pie chart represents native country and display percentages in legend respectively
df_data.columns = [i.replace('-', '_') for i in df_data.columns]
labels = df_data.native_country.unique()
sizes = df_data['native_country'].value_counts()
pie = sizes.plot(kind="pie", autopct='%1.1f%%',  figsize=(15, 10), shadow=True, legend=True, labeldistance=None)
plt.axis('equal')
percents = sizes.to_numpy() * 100 / sizes.to_numpy().sum()
plt.legend( bbox_to_anchor=(1.35,1.1), loc='upper right', labels=['%s, %1.1f %%' % (l, s) for l, s in zip(sizes.index,percents)])
plt.show()
```
